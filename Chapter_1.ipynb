$$
c
\begin{bmatrix}

4
\end{bmatrix}+
d
\begin{bmatrix}

3
\end{bmatrix}

\begin{bmatrix}

The length of that vector $\mathbf{w}$ is

14.

In Section 1.2, $\mathbf{v} \cdot \mathbf{w}$ will reveal the angle between those vectors.

$$

=
\begin{bmatrix}

\end{bmatrix}
=
\begin{bmatrix}

\end{bmatrix}.

$$
A

\end{bmatrix}

\begin{bmatrix}

\end{bmatrix}

And when we look at all combinations $Ax$ (with every $c$ and $d$), those vectors produce the column space of the matrix $A$. Here that column space is a plane.

With three vectors, the new matrix $B$ has columns $\mathbf{v}, \mathbf{w}, \mathbf{z}$.
In this example $\mathbf{z}$ is a combination of $\mathbf{v}$ and $\mathbf{w}$.
So the column space of $B$ is still the $xy$-plane.
The vectors $\mathbf{v}$ and $\mathbf{w}$ are independent but
$\mathbf{v}, \mathbf{w}, \mathbf{z}$ are dependent.
A combination produces zero:
$$
B
=
\begin{bmatrix}

\end{bmatrix}

\begin{bmatrix}

\end{bmatrix}

\text{ has}
$$

\begin{bmatrix}

-1
\end{bmatrix}

\begin{bmatrix}

4

0
\end{bmatrix}.
$$

The final goal is to understand matrix multiplication $AB = A$ times each column of $B$.

$$

#### 1.1.1 Linear Combinations $c\mathbf{v}+d\mathbf{w}$

Start from the beginning. A vector $\mathbf v$ in $2$-dimensional space has two components. To draw
$\mathbf v$ and $-\mathbf v$, use an arrow that begins at the zero vector:


```python
import numpy as np
import matplotlib.pyplot as plt

# Define vectors
v = np.array([2, 1])
w = np.array([2, -1])
v_plus_w = v + w
minus_v = -v
origin = np.array([0, 0])

plt.figure()

# Solid vectors from origin
plt.quiver(*origin, *v, angles='xy', scale_units='xy', scale=1)
plt.quiver(*origin, *w, angles='xy', scale_units='xy', scale=1)


# Resultant vector v + w
plt.quiver(*origin, *v_plus_w, angles='xy', scale_units='xy', scale=1)

# Negative vector -v
plt.quiver(*origin, *minus_v, angles='xy', scale_units='xy', scale=1,
           linestyle='dashed')

# Labels
plt.text(v[0]+0.05, v[1]+0.05, r"$\mathbf{v}$")
plt.text(w[0]+0.05, w[1]-0.2, r"$\mathbf{w}$")
plt.text(v_plus_w[0]+0.05, v_plus_w[1]+0.05, r"$\mathbf{v}+\mathbf{w}$")
plt.text(minus_v[0]-0.6, minus_v[1]-0.2, r"$-\mathbf{v}$")
plt.text(0.05, 0.05, r"$\mathbf{0}$")

# Axes
plt.axhline(0)
plt.axvline(0)
plt.gca().set_aspect('equal', adjustable='box')
plt.xlim(-3, 5)
plt.ylim(-3, 3)

plt.title("Vector addition: parallelogram law")
plt.show()
```

The vectors $c\mathbf{v}$ (for all numbers $c$) fill an infinitely long line
in the $xy$ plane. If $\mathbf{w}$ is not on that line, then the vectors
$d\mathbf{w}$ fill a second line. We aim to see that the linear combinations
$c\mathbf{v} + d\mathbf{w}$ fill the plane. Combining points on the
$\mathbf{v}$ line and the $\mathbf{w}$ line gives all points.

Here are four different linear combinations—we can choose any numbers $c$ and $d$:

$$
\begin{aligned}
1\mathbf{v} + 1\mathbf{w} &= \text{sum of vectors},\\
1\mathbf{v} - 1\mathbf{w} &= \text{difference of vectors},\\
0\mathbf{v} + 0\mathbf{w} &= \text{zero vector},\\
c\mathbf{v} + 0\mathbf{w} &= \text{vector } c\mathbf{v} \text{ in the direction of } \mathbf{v}.
\end{aligned}
$$


#### 1.1.2 Solving Two Equations

Solve

$$
c
\begin{bmatrix}
2 \\
1
\end{bmatrix}
+
d
\begin{bmatrix}
2 \\
-1
\end{bmatrix}
=
\begin{bmatrix}
8 \\
2
\end{bmatrix}.
$$

where $\mathbf{v}=\begin{bmatrix}2 \\1\end{bmatrix}$ and $\mathbf{w}=\begin{bmatrix}2 \\-1\end{bmatrix}$.

This means
$$
\begin{aligned}
2c + 2d &= 8,\\
c - d &= 2.
\end{aligned}
$$

Therefore, $c=3$ and $d=1$.

In matrix form
$$
\begin{bmatrix}
2 & 2 \\
1 & -1
\end{bmatrix}
\begin{bmatrix}
3 \\
1
\end{bmatrix}
=
\begin{bmatrix}
8 \\
2
\end{bmatrix}.
$$

#### Column Way, Row Way, Matrix Way

- **Column way**

    Linear combination:

    $
    c\begin{bmatrix}
    v_1 \\
    v_2
    \end{bmatrix}+d
    \begin{bmatrix}
    w_1 \\
    w_2
    \end{bmatrix}=
    \begin{bmatrix}
    b_1 \\
    b_2
    \end{bmatrix}
    $

- **Row way**  

    Two equations for $c$ and $d$:

    $
    \begin{aligned}
    v_1 c + w_1 d &= b_1 \\
    v_2 c + w_2 d &= b_2
    \end{aligned}
    $

- **Matrix way**  

    $2$ by $2$ matrix:

    $
    \begin{bmatrix}
    v_1 & w_1 \\
    v_2 & w_2
    \end{bmatrix}
    \begin{bmatrix}
    c \\
    d
    \end{bmatrix}=
    \begin{bmatrix}
    b_1 \\
    b_2
    \end{bmatrix}
    $

- Column view: $A\mathbf x$ is a linear combination of columns.
- Row view: $A\mathbf x=\mathbf b$ is a system of equations.
- Matrix view: $A\mathbf x=\mathbf b$.

If the points $\mathbf v$ and $\mathbf w$ and the zero vector $\mathbf 0$ are not on the same line, then we have:
- Unique solution $c$ and $d$.
- $\operatorname{Col}(A) = \mathbb{R}^2$ (the linear combinations of $\mathbf v$ and $\mathbf w$ exactly fill the $xy$ plane.)
- The vectors $\mathbf v$ and $\mathbf w$ are **Linear Independent**.
    - $c\mathbf v+d\mathbf w=\mathbf 0\Rightarrow c=d=0$
- The $2$ by $2$ matrix $A=\left[\mathbf v,\mathbf w\right]$ is invertible
    - The equation $A\mathbf x = \mathbf b$ has a unique solution $\mathbf x$ for every $\mathbf b\in\mathbb R^2$.
    - The determinant of $A$: $det(A)\neq 0$.

#### 1.1.3 Can Elimination Fail?

Elimination (Gaussian Elimination) fails happen when the vectors $\mathbf v$ and $\mathbf w$ lie on the same line through the center point $(0,0)$. Those vectors are not independent.

The reason is clear: All combinations of $\mathbf{v}$ and $\mathbf{w}$ will then lie on that same line.
If the desired vector $\mathbf{b}$ is off the line, then the equations $c\mathbf{v} + d\mathbf{w} = \mathbf{b}$ have no solution:

**Example**

0
\end{bmatrix}

$$

#### 1.1.4 Vectors in Three Dimensions

We need *three* independent vectors if we want their combinations to fill
3-dimensional space. Here is the simplest choice $\mathbf{i}, \mathbf{j}, \mathbf{k}$ for those three independent vectors:

$$
\mathbf{i} =
\begin{bmatrix}
1\\
0\\
0
\end{bmatrix},
\qquad
\mathbf{j} =
\begin{bmatrix}
0\\
1\\
0
\end{bmatrix},
\qquad
\mathbf{k} =
\begin{bmatrix}
0\\
0\\
1
\end{bmatrix}
$$

$$
c\mathbf{i} + d\mathbf{j} + e\mathbf{k}
=
\begin{bmatrix}
c\\
d\\
e
\end{bmatrix}
\tag{1}
$$

Now $\mathbf{i}, \mathbf{j}, \mathbf{k}$ go along the $x, y, z$ axes in
three-dimensional space $\mathbb{R}^3$. We can easily write any vector $\mathbf{v}$ in $\mathbb{R}^3$ as a combination of $\mathbf{i}, \mathbf{j}, \mathbf{k}$:

**Vector form**

$$
\mathbf{v} =
\begin{bmatrix}
v_1\\
v_2\\
v_3
\end{bmatrix}
=
v_1 \mathbf{i} + v_2 \mathbf{j} + v_3 \mathbf{k}
$$

**Matrix form**

$$
\begin{bmatrix}
1 & 0 & 0\\
0 & 1 & 0\\
0 & 0 & 1
\end{bmatrix}
\begin{bmatrix}
v_1\\
v_2\\
v_3
\end{bmatrix}
=
\begin{bmatrix}
v_1\\
v_2\\
v_3
\end{bmatrix}
$$

That is the $3$ by $3$ identity matrix $I$. Multiplying by $I$ leaves every vector unchanged. $I$ is the matrix analog of the number $1$, because $I\mathbf{v} = \mathbf{v}$ for every $\mathbf{v}$.

Notice that we could not take a linear combination of a 2-dimensional vector $\mathbf{v}$ and a 3-dimensional vector $\mathbf{w}$. They are not in the same space.


### 1.2 Lengths and Angles from Dot Products

1. The *dot product* of
$
\mathbf{v} =
\begin{bmatrix}
1\\
2
\end{bmatrix}
\quad \text{and} \quad
\mathbf{w} =
\begin{bmatrix}
4\\
6
\end{bmatrix}
$
is
$
\mathbf{v}\cdot\mathbf{w}
= (1)(4) + (2)(6)
= 4 + 12
= 16.
$

2. The length squared of
$
\mathbf{v} = (1,3,2)
$
is
$
\mathbf{v}\cdot\mathbf{v}
= 1 + 9 + 4
= 14.
$
The length is
$
\|\mathbf{v}\| = \sqrt{14}.
$

3. The vector
$
\mathbf{v} = (1,3,2)
$
is perpendicular to
$
\mathbf{w} = (4,-4,4)
$
because
$
\mathbf{v}\cdot\mathbf{w} = 0.
$

4. The angle $\theta = 45^\circ$ between
$
\mathbf{v} =
\begin{bmatrix}
1\\
0
\end{bmatrix}
\quad \text{and} \quad
\mathbf{w} =
\begin{bmatrix}
1\\
1
\end{bmatrix}
$
has
$
\cos\theta
=
\frac{\mathbf{v}\cdot\mathbf{w}}{\|\mathbf{v}\|\;\|\mathbf{w}\|}
=
\frac{1}{(1)(\sqrt{2})}.
$

5. All angles have
$
|\cos\theta| \le 1.
$
All vectors have
$
|\mathbf{v}\cdot\mathbf{w}| \le \|\mathbf{v}\|\;\|\mathbf{w}\|
\quad \text{(Schwarz inequality)}
$
and
$
\|\mathbf{v}+\mathbf{w}\| \le \|\mathbf{v}\| + \|\mathbf{w}\|
\quad \text{(Triangle inequality)}.
$

The most useful multiplication of vectors $\mathbf{v}$ and $\mathbf{w}$ is
their dot product $\mathbf{v}\cdot\mathbf{w}$. We multiply the first components $v_1 w_1$ and the second components $v_2 w_2$ and so on. Then we add those results to get a single number $\mathbf{v}\cdot\mathbf{w}$:

$$
\boxed{
\text{The dot product of }
\mathbf{v} =
\begin{bmatrix}
v_1\\
v_2
\end{bmatrix}
\text{ and }
\mathbf{w} =
\begin{bmatrix}
w_1\\
w_2
\end{bmatrix}
\text{ is }
\mathbf{v}\cdot\mathbf{w} = v_1 w_1 + v_2 w_2.
}
\tag{1}
$$

If the vectors are in $n$-dimensional space with $n$ components each, then

$$
\boxed{
\text{Dot product}\qquad
\mathbf{v}\cdot\mathbf{w}
=
v_1 w_1 + v_2 w_2 + \cdots + v_n w_n
=
\mathbf{w}\cdot\mathbf{v}.
}
\tag{2}
$$

The dot product $\mathbf{v}\cdot\mathbf{v}$ tells us the squared length $\|\mathbf{v}\|^2=v_1^2 + \cdots + v_n^2$ of a vector.

A unit vector $\mathbf{u}$ has length $\|\mathbf{u}\| = 1$.
If $\mathbf{v} \neq \mathbf{0}$ then
$
\mathbf{u} = \frac{\mathbf{v}}{\|\mathbf{v}\|}
$
is a unit vector.

**Example 1**  
The standard unit vector along the $x$ axis is written $\mathbf{i}$.
In the $xy$ plane, the unit vector that makes an angle $\theta$ with the
$x$ axis is
$
\mathbf{u} = (\cos\theta,\sin\theta).
$


$$
\text{Unit vectors:}\quad
\mathbf{i} =
\begin{bmatrix}
1\\
0
\end{bmatrix}
\qquad \text{and} \qquad
\mathbf{u} =
\begin{bmatrix}
\cos\theta\\
\sin\theta
\end{bmatrix}.
$$

Notice
$
\mathbf{i}\cdot\mathbf{u} = \cos\theta.
$ The vector $\mathbf{u} = (\cos\theta,\sin\theta)$ is a unit vector because
$\mathbf{u}\cdot\mathbf{u}=\cos^2\theta + \sin^2\theta=1$.




#### 1.2.1 Perpendicular Vectors

Suppose the angle between $\mathbf{v}$ and $\mathbf{w}$ is $90^\circ$.
Its cosine is zero. That produces a valuable test
$\mathbf{v}\cdot\mathbf{w} = 0$ for perpendicular vectors.

$$
\boxed{
\text{Perpendicular vectors have } \mathbf{v}\cdot\mathbf{w} = 0.
\text{ Then }
\|\mathbf{v}+\mathbf{w}\|^2 = \|\mathbf{v}\|^2 + \|\mathbf{w}\|^2.
}
\tag{3}
$$

This is the most important special case. It has brought us back to $90^\circ$ angles and lengths $a^2+b^2=c^2$. The algebra for perpendicular vectors $(\mathbf{v}\cdot\mathbf{w} = 0 = \mathbf{w}\cdot\mathbf{v})$
is easy:

$$
\begin{aligned}
\|\mathbf{v}+\mathbf{w}\|^2
&= (\mathbf{v}+\mathbf{w})\cdot(\mathbf{v}+\mathbf{w}) \\
&= \mathbf{v}\cdot\mathbf{v}
+ \mathbf{v}\cdot\mathbf{w}
+ \mathbf{w}\cdot\mathbf{v}
+ \mathbf{w}\cdot\mathbf{w} \\
&= \|\mathbf{v}\|^2 + \|\mathbf{w}\|^2.
\end{aligned}
\tag{4}
$$

Two terms were zero. Please notice that
$
\|\mathbf{v}-\mathbf{w}\|^2
$
is also equal to
$
\|\mathbf{v}\|^2 + \|\mathbf{w}\|^2.
$

**Example 2**

- The vector $\mathbf{v} = (1,1)$ is at a $45^\circ$ angle with the $x$ axis.  

- The vector $\mathbf{w} = (1,-1)$ is at a $-45^\circ$ angle with the $x$ axis.  

- The sum $\mathbf{v}+\mathbf{w}$ is $(2,0)$. The difference $\mathbf{v}-\mathbf{w}$ is $(0,2)$.

So the angle between $(1,1)$ and $(1,-1)$ is $90^\circ$. Their dot product is
$
\mathbf{v}\cdot\mathbf{w} = 1 - 1 = 0.
$

This right triangle has
$
\|\mathbf{v}\|^2 = 2,
\qquad
\|\mathbf{w}\|^2 = 2,
\qquad
\|\mathbf{v}-\mathbf{w}\|^2 = \|\mathbf{v}+\mathbf{w}\|^2 = 4.
$

**Example 3**

The vectors
$
\mathbf{v} = (4,2)
\quad \text{and} \quad
\mathbf{w} = (-1,2)
$
have a zero dot product:

**Dot product is zero** so **Vectors are perpendicular**

$$
\begin{bmatrix}
4\\
2
\end{bmatrix}
\cdot
\begin{bmatrix}
-1\\
2
\end{bmatrix}
=
-4 + 4
=
0.
$$



```python
import numpy as np
import matplotlib.pyplot as plt

# Angle (radians) - change as you like
theta = np.deg2rad(35)

# Unit circle
t = np.linspace(0, 2*np.pi, 600)
xc, yc = np.cos(t), np.sin(t)

# Unit vector u
u = np.array([np.cos(theta), np.sin(theta)])

plt.figure()

# Circle and axes
plt.plot(xc, yc)
plt.axhline(0)
plt.axvline(0)

# Basis vectors i, j
plt.quiver(0, 0, 1, 0, angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, 0, 1, angles='xy', scale_units='xy', scale=1)

# Vector u
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1)

# Projections (dashed)
plt.plot([0, u[0]], [0, 0], linestyle='dashed')
plt.plot([u[0], u[0]], [0, u[1]], linestyle='dashed')

# Angle arc
arc = np.linspace(0, theta, 200)
plt.plot(0.2*np.cos(arc), 0.2*np.sin(arc))

# Labels
plt.text(1.05, -0.05, r"$\mathbf{i}$")
plt.text(-0.08, 1.05, r"$\mathbf{j}$")
plt.text(u[0] + 0.05, u[1] + 0.05, r"$\mathbf{u}$")
plt.text(u[0]/2, -0.12, r"$\cos\theta$")
plt.text(u[0] + 0.05, u[1]/2, r"$\sin\theta$")
plt.text(0.22*np.cos(theta/2), 0.22*np.sin(theta/2), r"$\theta$")
plt.text(-0.35, -1.18, "Unit circle")

# Formatting
plt.gca().set_aspect('equal', adjustable='box')
plt.xlim(-1.3, 1.3)
plt.ylim(-1.3, 1.3)
plt.title(r"Unit circle with $\mathbf{u}=(\cos\theta,\sin\theta)$")
plt.show()
```

#### 1.2.2 A Plane in 3 Dimensions

Suppose $\mathbf n$ is a unit vector with three components $n_1,n_2,n_3$. Look at all vectors $\mathbf w$ in $\mathbb R^3$ that are perpendicular to $\mathbf n$ (so $\mathbf w\cdot \mathbf n = 0$):

$$
\text{The vectors } \mathbf{w} \text{ with } \mathbf{w}\cdot\mathbf{n}=0
\text{ fill a 2-dimensional plane in } \mathbb{R}^3.

Section 1.3 begins the algebra of $m$ by $n$ matrices: our true goal. A typical matrix $A$ is a rectangle of $m$ times $n$ numbers — $m$ rows and $n$ columns. If $m$ equals $n$ then $A$ is a *square matrix*. The examples below are $3$ by $3$ matrices.

We often think of the columns of $A$ as vectors $\mathbf{a}_1, \mathbf{a}_2, \ldots, \mathbf{a}_n$. Each of those $n$ vectors is in $m$-dimensional space. In this example the $\mathbf{a}$’s have $m = 3$ components each:

$$

We often think of the columns of $A$ as vectors $\mathbf{a}_1, \mathbf{a}_2, \ldots, \mathbf{a}_n$. Each of those $n$ vectors is in $m$-dimensional space. In this example the $\mathbf{a}$’s have $m = 3$ components each:

$$
m = 3 \text{ rows}
$$
$$
n = 4 \text{ columns}
$$

$$
\text{$3$ by $4$ matrix } A
=

\end{bmatrix}
=
\begin{bmatrix}

0 & 0 & -1 & 1
\end{bmatrix}.
$$

This is a *difference matrix* because $A$ times $\mathbf{x}$ produces a vector $A\mathbf{x}$ of differences like $x_2 - x_1$. How does an $m$ by $n$ matrix $A$ multiply an $n$ by $1$ vector $\mathbf{x}$? There are two ways to the same answer—we work with the rows of $A$
or we work with the columns.

$$
\boxed{
\begin{aligned}
\text{The row picture of } A\mathbf{x}
&\text{ will come from dot products of } \mathbf{x}
\text{ with the rows of } A.\\
\text{The column picture}
&\text{ will come from linear combinations of the columns of } A.
\end{aligned}
}
$$

The next important words are **independence, dependence, and column space**.

1. **Independence:**

    Independence means: The only combination of columns that produces $A\mathbf{x}=(0,0,0)$ is $\mathbf{x}=(0,0,0)$. The columns are independent when each new column is a vector that we don’t already have as a combination of previous columns.
2. **Dependence:**

    The opposite of independent is **dependen**. For example: **Dependent columns**

    $$
    A_2 =
    \begin{bmatrix}
    1 & 2 & 3\\
    1 & 4 & 5\\
    6 & 0 & 6
    \end{bmatrix}
    $$

    Column 1 + Column 2 = Column 3: **Their combinations don’t fill 3D space**
    
    $$
    \begin{aligned}
    1 + 2 &= 3,\\
    1 + 4 &= 5,\\
    6 + 0 &= 6.
    \end{aligned}
    $$
3. **Column Space:**

    The column space $C(A)$ contains all vectors $A\mathbf{x}$: all combinations of the columns.

    $$C(A)=\left\{A\mathbf{x}:\mathbf{x}\in\mathbb{R}^n\right\}=\left\{x_1\mathbf{a}_1+x_2\mathbf{a}_2+\cdots+x_n\mathbf{a}_n\right\}.$$

#### 1.3.1 Thinking About the Column Space of $A$

For $m = 3$, here are all four possibilities for column spaces in $3$-dimensional space:

   $A$ has $3$ independent columns.

2. **A plane in $\mathbb{R}^3$ going through $(0,0,0)$**  
   $A$ has $2$ independent columns.

3. **A line in $\mathbb{R}^3$ going through $(0,0,0)$**  
   $A$ has $1$ independent column.

4. **The single point $(0,0,0)$ in $\mathbb{R}^3$**  
   $A$ is a matrix of zeros.

$

$
\begin{bmatrix}

\end{bmatrix}

$

$

Here is a succession of questions. With practice in the next section 1.4, you will find the keys to the answers. They give a real understanding of any matrix $A$.

1. **How many columns of $A$ are independent?**  
   That number $r$ is the *rank* of $A$.

2. **Which are the first $r$ independent columns?**  
   They are a *basis* for the column space.

3. **What combinations of those $r$ basic columns produce the remaining
   $n-r$ columns?**

4. **Write any $A$ as an $m$ by $r$ column matrix $C$ times an $r$ by $n$
   matrix $R$:** $A = CR.$

5. **(Amazing)** The $r$ rows of $R$ are a basis for the *row space* of $A$: combinations of rows.


#### 1.3.2 Matrices of Rank One

**Example**

$

\begin{bmatrix}

4 & 12 & -8\\

\end{bmatrix}
$ has rank $r = 1$.


Here is a wonderful fact about any rank one matrix. You may have noticed the rows of $A_6$. **All the rows are multiples of one row. When the column space is a single line in $m$-dimensional space, the row space is a single line in $n$-dimensional space.**

$
A =
\begin{bmatrix}
a & ma & pa\\
b & mb & pb\\
c & mc & pc
\end{bmatrix}
$

- Column $2$ is $m$ times column $1$.  

- Column $3$ is $p$ times column $1$.  

- Rows $2$ and $3$ are $\dfrac{b}{a}$ and $\dfrac{c}{a}$ times row $1$.

It looks possible that **row rank equals column rank for every matrix. If $A$ has $r$ independent columns, then $A$ has $r$ independent rows.** A wonderful fact! The next section 1.4 will explain matrix multiplication $AB$ and lead us toward an understanding of “row rank = column rank” for all matrices.

### 1.4 Matrix Multiplication $AB$ and $CR$

1. To multiply $AB$ we need row length for $A$ = column length for $B$.

2. The number in row $i$, column $j$ of $AB$ is (row $i$ of $A$) $\cdot$ (column $j$ of $B$).

3. **By columns:** $A$ times column $j$ of $B$ produces column $j$ of $AB$.

4. Usually $AB$ is different from $BA$. But always $(AB)C = A(BC).$

5. If $A$ has $r$ independent columns in $C$, then $A = CR = (m \times r)(r \times n).$

This section moves to matrix-matrix multiplication : a matrix $A$ times a matrix $B$. The new rule builds on the old one, when the matrix $B$ has columns $b_1, b_2,...,b_p$. We just multiply $A$ times each of those $p$ columns of $B$ to find the $p$ columns of $AB$.

$$\boxed{
\begin{aligned}
&\text{Column } j \text{ of } AB \text{ equals } A \text{ times column } j \text{ of } B\\&\text{If } B = \begin{bmatrix}\mathbf{b}_1 \ \cdots \ \mathbf{b}_p\end{bmatrix} \text{ then }AB =\begin{bmatrix}A\mathbf{b}_1 \ \cdots \ A\mathbf{b}_p\end{bmatrix}\end{aligned}
}\tag{1}$$

For a 2 by 2 **exchange matrix** $B$:

$$
A\mathbf{b}_1
=
\begin{bmatrix}

3 & 4
\end{bmatrix}
\begin{bmatrix}

1
\end{bmatrix}
=
\begin{bmatrix}

4
\end{bmatrix}
\quad

=
\begin{bmatrix}

\end{bmatrix}
\begin{bmatrix}

\end{bmatrix}
=
\begin{bmatrix}

\end{bmatrix}
\quad

=
\begin{bmatrix}

\end{bmatrix}
\begin{bmatrix}

\end{bmatrix}
=
\begin{bmatrix}

\end{bmatrix}
$$

For this matrix $B$, the result of multiplying $AB$ is to exchange the columns of $A$.

- **Row way**  
    $mp$ dot products in $AB$, $n$ multiplications each: $mnp$ small multiplications.

- **Column way**  
    $p$ columns in $AB$, $mn$ multiplications each: $mnp$ small multiplications.

#### $AB$ is usually different from $BA$

For
$
B =
\begin{bmatrix}
0 & 1\\
1 & 0
\end{bmatrix},
$, $AB$ exchanges the columns of $A$. **But $BA$ exchanges the rows of $A$!**

$$
AB =
\begin{bmatrix}
2 & 1\\
4 & 3
\end{bmatrix}
\quad
BA =
\begin{bmatrix}
0 & 1\\
1 & 0
\end{bmatrix}
\begin{bmatrix}

\end{bmatrix}
=
\begin{bmatrix}

\end{bmatrix}.
\tag{6}
$$

Matrix multiplication is **not commutative**. In general, $BA \neq AB$ . Multiply $A$ on the **left** for row operations on $A$, and multiply on the **right** by $B$ for column operations on $A$.

**Squaring the exchange matrix give identity matrix:**

$$
B^2 =
\begin{bmatrix}
0 & 1\\
1 & 0
\end{bmatrix}
\begin{bmatrix}
0 & 1\\
1 & 0
\end{bmatrix}=
I
$$

For matrix multiplication, **this associative law is true**.

We can multiply $AB$ first or we can multiply $BC$ first.

The matrices stay in the order $A, B, C$ and their sizes must be right for multiplication:

$
A \text{ is } m \times n, 
\quad B \text{ is } n \times p, 
\quad C \text{ is } p \times q.
$ 

Then $AB \text{ is } m \times p$ and $(AB)C \text{ is } m \times q.$

We can test the law using the exchange matrix $B$ on the rows and the columns of $A$:

$$

\end{bmatrix}

\end{bmatrix}
\begin{bmatrix}

1 & 0
\end{bmatrix}
=
\begin{bmatrix}

2 & 1
\end{bmatrix}
$$

$$

=
\begin{bmatrix}

\end{bmatrix}
\begin{bmatrix}

\end{bmatrix}

\end{bmatrix}
$$

$$\boxed{
\textbf{Associative } (AB)C = A(BC)
\quad \text{and} \quad
\textbf{Distributive } A(B + C) = AB + AC
}
\tag{7}$$

#### Review of AB

- **Dot products**  

    $
    (\text{Row } i \text{ of } A)\cdot(\text{Col } j \text{ of } B)

    $

- **Combine columns**  

    $
    (\text{Matrix } A)(\text{Column } b_j \text{ of } B)
    = \text{vector in column } j \text{ of } AB
    $

The columns $Ab_j$ in $AB$ are combinations of columns of $A$.


#### Rank One Matrices and $A = CR$

Only one independent row when there is only one independent column. Our approach is through matrix multiplication. We factor $A$ into $C$ times $R$. For this very special matrix, $C$ has one column and $R$ has one row. $CR$ is $(3\times 1)(1\times 4)$.

$$\text{Rank} = 1
\qquad
\boxed{
A =
\begin{bmatrix}
1 & 2 & 10 & 100 \\
3 & 6 & 30 & 300 \\
2 & 4 & 20 & 200
\end{bmatrix}
=
\begin{bmatrix}
1 \\
3 \\
2
\end{bmatrix}
\begin{bmatrix}
1 & 2 & 10 & 100
\end{bmatrix}
= CR}
\tag{8}$$

$$\boxed{\text{If the column space of }A\text{ is a line, the row space of }A\text{ is also a line.}}$$

One column in $C$, one row in $R$. **Our next goal is to allow $r$ columns in $C$ and to find $r$ rows in $R$. And to see $A = CR$. That number $r$ is the “rank” of $A$.**


#### $C$ Contains the First $r$ Independent Columns of $A$

Suppose we go from left to right, looking for independent columns in any matrix $A$:

$$
\boxed{
\begin{aligned}
&\text{If column 1 of } A \text{ is not all zero, put it into the matrix } C. \\
&\text{If column 2 of } A \text{ is not a multiple of column 1, put it into } C. \\
&\text{If column 3 of } A \text{ is not a combination of columns 1 and 2, put it into } C. \\
&\text{Continue.}
\end{aligned}
}
$$

At the end, $C$ will have $r$ columns taken from $A$. That number $r$ is the **rank** of $A$ and $C$. The $n$ columns of $A$ might be dependent. The $r$ columns of $C$ will surely be **independent**.

**Independent columns**
- No column of $C$ is a combination of previous columns.
- No combination of columns gives $Cx = 0$ except $x = 0$.

Those $r$ independent columns in $C$ combine to give all $n$ columns in $A$.

**Example 2**

$$
A =
\begin{bmatrix}
2 & 6 & 4 \\
4 & 12 & 8 \\
1 & 3 & 5
\end{bmatrix}
\qquad \text{leads to} \qquad
C =
\begin{bmatrix}
2 & 4 \\
4 & 8 \\
1 & 5
\end{bmatrix}
\qquad
\text{Rank } r = 2
$$

Columns 1 and 3 go into $C$. Column 2 is $3$ times column 1: *not independent, not in $C$.*



#### Matrix Multiplication $C$ times $R$

The first column of $A$ is actually in $C$, so the first column of $R$ just has $1$ and $0$. The third column of $A$ comes second in $C$, so the third column of $R$ just has $0$ and $1$.

$$
\textbf{Notice } I \text{ inside } R
\qquad
\text{Rank } r = 2
\quad
A = CR
\quad\text{is}\quad
\begin{bmatrix}
2 & 6 & 4 \\
4 & 12 & 8 \\
1 & 3 & 5
\end{bmatrix}=
\begin{bmatrix}
2 & 4 \\
4 & 8 \\
1 & 5
\end{bmatrix}
\begin{bmatrix}
1 & ? & 0 \\
0 & ? & 1
\end{bmatrix}
\tag{9}
$$

Two columns of $A$ went straight into $C$, so part of $R$ is the identity matrix. The question marks are in column 2 because column 2 of $A$ is **not** in $C$. It is a dependent column. Column 2 of $A$ is $3$ times column 1, so that number $3$ goes into $R$.

$$
A \text{ is } m \times n,
\quad
C \text{ is } m \times r,
\quad
R \text{ is } r \times n
\quad
\boxed{
A = CR
\quad\text{is}\quad
\begin{bmatrix}
2 & 6 & 4 \\
4 & 12 & 8 \\
1 & 3 & 5
\end{bmatrix}
=
\begin{bmatrix}
2 & 4 \\
4 & 8 \\
1 & 5
\end{bmatrix}
\begin{bmatrix}
1 & 3 & 0 \\
0 & 0 & 1
\end{bmatrix}}
\tag{10}
$$

That example is typical of $A = CR$. We review the descriptions of $C$ and $R$.

1. $C$ contains a full set of $r$ **independent columns** (chosen left to right) in $A$.
2. $R = [\, I \;\; F \,]$ contains the identity matrix $I$ in the same $r$ columns that held $C$.
3. The **dependent columns** of $A$ are combinations $CF$ of the independent columns in $C$.

That matrix $F$ goes into the other $n-r$ columns of  
$
R = [\, I \;\; F \,].
$

$
A = C[\, I \;\; F \,]= [\, C \;\; CF \,]= [\, \text{indep cols of } A \;\; \text{dep cols of } A \,]\quad (\text{in correct order})
$

$C$ has the same column space as $A$. $R$ has the same row space as $A$. Here
$
F =
\begin{bmatrix}
-1 \\
2
\end{bmatrix}
$

**Example 3 of $A = CR$ (Rank 2)**

$$
\begin{bmatrix}

\end{bmatrix}
=
\begin{bmatrix}

\end{bmatrix}
\begin{bmatrix}

\end{bmatrix}
\tag{11}
$$

When a column of $A$ goes into $C$, a column of $I$ goes into $R$.

$$
\boxed{
\text{Column } j \text{ of } A = C \times \text{column } j \text{ of } R
\qquad
\text{Row } i \text{ of } A = \text{row } i \text{ of } C \times R
}
$$

If all columns of $A$ are independent, then $C = A$. What matrix is $R$? **Answer:** $R = I$.

$$\boxed{
\begin{aligned}
&\text{Chapter 1 finds } C \text{ (independent columns of } A) \text{ before } R. \text{ Chapter 3 will find } R \text{ first.} \\
\\
&\text{Here column 3 of } A \text{ is the 2nd independent column in } C. \text{ Then column 3 of } R \text{ is }
\begin{bmatrix}
0 \\
1
\end{bmatrix}. \\
\\
&A =
\begin{bmatrix}

1 & 2 & 4 & 5
\end{bmatrix}
=
\begin{bmatrix}
1 & 3 \\
1 & 4
\end{bmatrix}
\begin{bmatrix}
1 & 2 & 0 & 1 \\
0 & 0 & 1 & 1
\end{bmatrix}
= CR\text{ All three ranks} = 2. \\
\\
&R \text{ tells how to recover all columns of } A \text{ from the independent columns in } C.
\end{aligned}
}$$


Here is an informal proof that the row rank of $A$ equals the column rank of $A$

1. The $r$ columns of $C$ are independent (they are chosen that way from $A$).

2. Every column of $A$ is a combination of those $r$ columns of $C$ (this is $A = CR$).

3. The $r$ rows of $R$ are independent (they contain the $r \times r$ identity matrix $I$).

4. Every row of $A$ is a combination of the $r$ rows of $R$ (this is $A = CR$ by rows).


#### How to Find the Matrix $R$

Up to now you have had very little help in discovering the matrix $R$ in  
$$

$$

If you could tell that column 3 of this matrix $A$ is a combination of columns 1 and 2, then the numbers $x$ and $y$ in that combination will go into column 3 of $R$:

**Example 4**

$$

\end{bmatrix}

\begin{bmatrix}

\end{bmatrix}

\end{bmatrix}

But even for this small matrix, we can’t immediately see $x$ and $y$.  
So we don’t know the rank of $A$ ($2$ or $3$?). There has to be a good way to discover $x$ and $y$.

That good way is **elimination**. It will be the key algorithm in Chapter 2 for square matrices and again in Chapter 3 for all matrices. We want to introduce it now for this matrix.

The idea is to simplify $A$ by **row operations**. That will simplify the equations for $x$ and $y$.

We will eliminate the $2$ and $3$ in column 1 of $A$. To do that, subtract $2$ times row 1 from row 2 of $A$, and also subtract $3$ times row 1 from row 3. The matrix $A$ changes to $B$:

$$

\begin{bmatrix}

\end{bmatrix}
\quad

\begin{bmatrix}

0 \\
0
\end{bmatrix}
+

\begin{bmatrix}

-2 \\
-2
\end{bmatrix}
\; ?=\;
\begin{bmatrix}

-6 \\
-6
\end{bmatrix}
\tag{13}
$$

We only did what is legal. *Subtracting an equation from an equation leaves a new equation.*

The new equation is
$$
-2y = -6,
$$
so we know
$$
y = 3.
$$

Then if
$$
x = -5,
$$
the top equation becomes
$$
-5 + 9 = 4,
$$
which is correct.

The original equations (12) are solved by $(-5, 3)$:

$$

\quad

\begin{bmatrix}

\end{bmatrix}

\begin{bmatrix}

\end{bmatrix}

\end{bmatrix}
\quad
\textbf{Column 3 of } A \textbf{ is dependent.}$$

So $-5$ and $3$ are the numbers we needed in column 3 of $R$.  
All the ranks are $r = 2$:

$$
A =
\begin{bmatrix}
1 & 3 & 4 \\
2 & 4 & 2 \\
3 & 7 & 6
\end{bmatrix}
=
\begin{bmatrix}
1 & 3 \\
2 & 4 \\
3 & 7
\end{bmatrix}
\begin{bmatrix}
1 & 0 & -5 \\
0 & 1 & 3
\end{bmatrix}
= CR
\tag{14}
$$

There is more to see in this example. The elimination process that reduced $A$ to $B$ is called **row reduction**. I will complete it from $B$ to $U$, to make the matrix even simpler. Just subtract row 2 of $B$ from row 3 of $B$ to see a row of zeros in $U$:

$$
A \;\to\;
B =
\begin{bmatrix}
1 & 3 & 4 \\
0 & -2 & -6 \\
0 & -2 & -6
\end{bmatrix}
\;\to\;
U =
\begin{bmatrix}
1 & 3 & 4 \\
0 & -2 & -6 \\
0 & 0 & 0
\end{bmatrix}
=
\text{upper triangular matrix } U
\tag{15}
$$

That zero row is a clear signal: the **row rank** is also $2$.
- Chapter 2 will stop with $U$.  
- Chapter 3 will eliminate upward to produce more zeros.  
We end up with $R_0$ and $R$:

$$
U =
\begin{bmatrix}
1 & 3 & 4 \\
0 & -2 & -6 \\
0 & 0 & 0
\end{bmatrix}
\;\to\;
\begin{bmatrix}
1 & 3 & 4 \\
0 & 1 & 3 \\
0 & 0 & 0
\end{bmatrix}
\;\to\;
\begin{bmatrix}
1 & 0 & -5 \\
0 & 1 & 3 \\
0 & 0 & 0
\end{bmatrix}
= R_0
$$

$$
\boxed{
\begin{aligned}
&\text{All rows of } R_0 \text{ are combinations of the original rows of } A. \\
&\text{That zero row of } R_0 \text{ shows that } A \text{ has rank } r = 2. \\
&\text{The } 2 \times 2 \text{ identity matrix shows that columns } 1,2 \text{ of } A \text{ are independent (in } C). \\
&\text{Removing the zero row of } R_0 \text{ leaves the desired matrix } R \text{ in } A = CR. \\
&\text{Elimination in Chapter 3 will be a systematic way to find } R.
\end{aligned}
}
$$

$$
\boxed{
\begin{aligned}
\textbf{Key facts } A = CR: \qquad
&\text{The } r \text{ columns of } C \text{ are a basis for the column space of } A \ (\text{dimension } r). \\
&\text{The } r \text{ rows of } R \text{ are a basis for the row space of } A \ (\text{dimension } r).
\end{aligned}
}
$$

Those words **“basis”** and **“dimension”** will be properly defined later in Section 3.4.

- Chapter 1 starts with **independent columns of $A$**, placed in $C$.  
- Chapter 3 starts with the **rows of $A$**, and combines them into $R$.

We are emphasizing $CR$ because both matrices are so important. $C$ contains $r$ independent columns of $A$. $R$ tells how to combine those columns to give all columns of $A$ ($R$ contains $I$, because $r$ columns of $A$ are already in $C$). Chapter 3 will produce $R$ directly from $A$ by **elimination**, the most used algorithm in computational mathematics.

$$
A = CR \quad \text{will be the key to a fundamental problem: solving linear equations } Ax = b.
$$




#### Columns of $A$ times Rows of $B$ ... Columns of $C$ times Rows of $R$

Before this chapter ends, I want to add this message. There is another way to multiply matrices (producing the same matrix $AB$ or $CR$ as always). This way is not so well known, but it is powerful.

**The new way multiplies columns of $A$ times rows of $B$.**

$$
\boxed{
AB
=
\begin{bmatrix}
\,| &        & | \\
a_1 & \cdots & a_n \\
\,| &        & |
\end{bmatrix}
\begin{bmatrix}
- b_1^{*} - \\
\vdots \\
- b_n^{*} -
\end{bmatrix}
=
a_1 b_1^{*} + a_2 b_2^{*} + \cdots + a_n b_n^{*}}
\tag{16}
$$

Those matrices $a_k b_k^{*}$ are called **outer products**. We recognize that they have **rank one**: column times row. They are entirely different from **dot products** (rows times columns). If $A$ is an $m \times n$ matrix and $B$ is an $n \times p$ matrix, then columns of $A$ times rows of $B$ adds up to the same answer $AB$ as dot products of rows of $A$ and columns of $B$.

$$
AB \text{ involves the same } mnp \text{ small multiplications but in a new order!}
$$

$$
\begin{aligned}
(\text{Row})(\text{Column}) &: \quad mp \text{ dot products, } n \text{ multiplications each} 
\quad \Rightarrow \quad \text{total } mnp \\
(\text{Column})(\text{Row}) &: \quad n \text{ rank-one matrices, } mp \text{ multiplications each} 
\quad \Rightarrow \quad \text{total } mnp
\end{aligned}
$$

$$
\text{Columns × Rows}
\quad
\begin{bmatrix}
1 & 4 \\
2 & 5 \\
3 & 6
\end{bmatrix}
\begin{bmatrix}
7 & 8 & 9 \\
10 & 11 & 12
\end{bmatrix}
=
\begin{bmatrix}
1 \\
2 \\
3
\end{bmatrix}
[\,7\;\;8\;\;9\,]+
\begin{bmatrix}
4 \\
5 \\
6
\end{bmatrix}
[\,10\;\;11\;\;12\,]
$$

$$
\text{Rank 1 + Rank 1}
\quad
\begin{bmatrix}
7 & 8 & 9 \\
14 & 16 & 18 \\
21 & 24 & 27
\end{bmatrix}+
\begin{bmatrix}
40 & 44 & 48 \\
50 & 55 & 60 \\
60 & 66 & 72
\end{bmatrix}
=
\begin{bmatrix}
47 & 52 & 57 \\
64 & 71 & 78 \\
81 & 90 & 99
\end{bmatrix}
= AB
$$

This example has

$$
mnp = (3)(2)(3) = 18.
$$

At the start of the second line you see the **18 multiplications** (in two $3 \times 3$ matrices). Then **9 additions** give the correct answer $AB$.

As we learned in this section, the **rank of $AB$ is 2**. Two independent columns, not three. Two independent rows, not three.

The next chapter uses different words. $AB$ has **no inverse matrix**: it is **not invertible**. And in Chapter 5: **The determinant of $AB$ is zero.**


#### Note about the matrix $R$

We were amazed to learn that the row matrix $R$ in $A = CR$ is already a famous matrix in linear algebra! It is essentially the **reduced row echelon form** of the original $A$.

```python
import sympy as sp

A = sp.Matrix([
    [1, 3, 4],
    [2, 4, 2],
    [3, 7, 6],
])

R0, pivots = A.rref()   # R0 is RREF, pivots is tuple of pivot column indices
print("Pivot columns:", pivots)
print("RREF:")
sp.pprint(R0)
```

**Example 5**

$$
A =
\begin{bmatrix}
a_1 & a_2 & 3a_1 + 4a_2
\end{bmatrix}
=
\begin{bmatrix}
a_1 & a_2
\end{bmatrix}
\begin{bmatrix}
1 & 0 & 3 \\
0 & 1 & 4
\end{bmatrix}
= CR
$$

Here $a_1$ and $a_2$ are the **independent columns** of $A$. The third column is **dependent** — a combination of $a_1$ and $a_2$.  
Therefore it is in the plane produced by columns 1 and 2. All three matrices $A, C, R$ have rank $r = 2$.

We can try that new way (**columns × rows**) to quickly multiply $CR$ in Example 5:

$$
\text{Columns of } C \text{ times rows of } R
$$

$$
CR
=
a_1
\begin{bmatrix}
1 & 0 & 3
\end{bmatrix}
+
a_2
\begin{bmatrix}
0 & 1 & 4
\end{bmatrix}
=
\begin{bmatrix}
a_1 & a_2 & 3a_1 + 4a_2
\end{bmatrix}
= A
$$
